```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
gray_image = cv2.imread('./cameraman.png', 0)
canny_image = cv2.Canny(gray_image, 100, 200)

plt.figure(figsize = (10,10))
plt.imshow(canny_image, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
# Taking a matrix of size 5 as the kernel
kernel = np.ones((2, 2), np.uint8)
canny_dilated_image = cv2.dilate(canny_image, kernel, iterations = 1)

plt.figure(figsize = (10,10))
plt.imshow(canny_dilated_image, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
img_erosion = cv2.erode(canny_dilated_image, kernel, iterations=1)
plt.figure(figsize = (10,10))
plt.imshow(img_erosion, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('./cameraman.png',0)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
rows, cols = img.shape
crow,ccol = rows/2 , cols/2

fshift[int(crow-30):int(crow+30), int(ccol-30):int(ccol+30)] = 0
f_ishift = np.fft.ifftshift(fshift)
reconstruct_image = np.fft.ifft2(f_ishift)
reconstruct_image = np.abs(reconstruct_image)

plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(reconstruct_image, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(reconstruct_image)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])

plt.show()
```
